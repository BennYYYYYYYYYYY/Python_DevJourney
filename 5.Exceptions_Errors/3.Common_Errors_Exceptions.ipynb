### Some extremely common errors and exceptions should know about:
1. TypeError
2. ZeroDivisionError
3. NameError
4. RecursionError
5. KeyError
6. IndexError
7. ValueError
8. FileNotFoundError


## TypeError

```python
try:
    for i in range["a", "b", "c"]:
        print(sum(i))
except TypeError as e:
    print(e)
```

## ZeroDivisionError

```python
try:
    x = (lambda x,y : x / y)(1, 0)
    print(x)
except ZeroDivisionError as e:
    print(e)
```

## NameError

```python
try:
    print(x)
except NameError as e:
    print(e)
```

## RecursionError

```python
try:
    def hello():
        print("hello")
        hello() # stack overflow
    hello()
except RecursionError as e:
    print(e)
```

## KeyError 

```python
try:
    a = {1:"B", 2:"E", 3:"N", 4:"N", 5:"Y"}
    print(a[6])
except KeyError as e:
    print(e, "Something wrong here...")
```

## IndexError

```python
try:
    b = [1, 2, 3, 4, 5]
    print(b[5])
except IndexError as e:
    print(e)
```

## ValueError

```python
try:
    print(int("hello"))
except ValueError as e:
    print(e)
```

## FileNotFoundError

```python
try:
    with open("123") as f:
        print(f.read())
except FileNotFoundError as e:
    print(e)
```
