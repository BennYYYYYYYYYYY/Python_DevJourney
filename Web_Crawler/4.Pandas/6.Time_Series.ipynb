# datetime

```python
from datetime import datetime

time = datetime.now()
print(time)
```

```python
time_stop = datetime(2024, 11, 20, 9, 33)
while datetime.now() < time_stop:
    print('Program is sleeping.', end='')
print("Wake up")
```

```python
# Time Delta (一段時間)
from datetime import timedelta

delta = timedelta(days=0, seconds=5, microseconds=0, milliseconds=0, minutes=0, hours=0, weeks=0)
print(delta.total_seconds())
```

```python
# # Time Series 就是將 index 用日期取代

import pandas as pd
from datetime import datetime, timedelta

ndays = 5
start = datetime(2024, 11, 20)
dates = [start + timedelta(days=x) for x in range(0, ndays)]
data = [34, 44, 65, 53, 39]
time_series = pd.Series(data, index=dates) # index 用日期
print(type(time_series))
time_series
```

# Pandas 的時間區間方法

Pandas 的 `date_range()`可以產生時間區間。

```python
dates = pd.date_range("2024-11-10", "2024-11-20")
print(dates) # DatetimeIndex

data = [x for x in range(0, 11)]
ts = pd.Series(data, index=dates)
ts # "Freq: D" 表示索引是日期 
```

```python
print(pd.date_range(start="2024-11-1", periods=2)) # 日期往後
print(pd.date_range(end="2024-11-1", periods=3))  # 日期往前
print(pd.date_range("2024-11-1", "2024-11-20", freq='B')) # freq='B': 讓時間只包含workday，沒有包含假日 ，參數可以再查
```

```python
# 繪製 Time Series 折線圖
import random
import matplotlib.pyplot as plt

dates = pd.date_range('2024-11-15', '2024-11-20')
data = [random.random() for _ in range(6)]
ts = pd.Series(data, index=dates)
ts.plot(title="Time Series")
plt.xlabel("Date")
plt.ylabel("Data")
plt.show()
```
