```python
import requests

url = r"https://www.ccu.edu.tw/"
response = requests.get(url)

print(type(response)) # Response object 這個實例包含了從伺服器返回的所有響應數據(Bytes)
```

```python
# Response's attribute

response = requests.get(r"https://www.ccu.edu.tw/")
print(response.status_code,"\n") # 2xx 成功, 3xx 重定向, 4xx 客戶端錯誤, 5xx 伺服器錯誤  
print(response.ok) # boolean value, 當 response.status_code return <400 時 = True 
print(response.url,"\n") 
print(response.text,"\n") # 網頁內容 
print(response.content,"\n") # bytes 類型資料
print(response.headers,"\n")
print(response.cookies)
```

```python
# 尋找網頁特定內容

import re 
response = requests.get(r"https://www.ccu.edu.tw/")
if response.status_code == response.ok: # if return <400
    pattern = input("Keyword that looking for: ")

    name = re.findall(pattern, response.text) # return list of every match words
    if name != None:
        print(len(name))
    else:
        print("0")
else:
    print("Fail")
```

#### `raise_for_status()`  exception control

```python
response = requests.get(r"https://www.ccu.edu.tw/nonexistentpage")
try:
    response.raise_for_status() # 狀態碼為 4xx 或 5xx 時引發一個例外 (HTTPError)
    print("Download success")
except Exception as e:
    print(e)
```

#### 偽裝瀏覽器 `headers`

```python
headers = { 'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64)\
            AppleWebKit/537.36 (KHTML, like Gecko) Chrome/45.0.2454.101\
            Safari/537.36', }
url = 'https://www.kingstone.com.tw/basic/2013120599512?zone=book&lid=search&actid=WISE'
response = requests.get(url, headers=headers)
response.raise_for_status()
print("偽裝瀏覽器擷取網路資料成功")
```

#### `iter_content()` 逐塊(chunk)讀取 HTTP 回應內容，文件必須以二進位模式寫入，確保數據的完整性。

```python
url = 'https://github.com/BennYYYYYYYYYYY?tab=repositories'
response = requests.get(url, stream=True) # stream=True 以啟用分塊讀取(文件流)

with open('largefile.zip', 'wb') as f: # wb (write bytes)
    for chunk in response.iter_content(chunk_size=1024):
        if chunk:  # 避免寫入空塊
            f.write(chunk)
```
